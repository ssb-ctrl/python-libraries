```python
a1 = np.arange(1,11)
a2 = np.arange(11,21)
print(a1+a2)
print(a1-a2)
print(a1*a2)
print(a1**a2)
print(a2/a1)
print(a2//a1)
```

```python
#broadcasting
arr = np.arange(1,21)
print(arr +10)
print(arr-10)
print(arr*10)
print(arr**2)
print(arr/10)
print(arr//10)
# these operations can be performed for 2d array as well
```

```python
#deep copy and shallow copy 
"""
1)shallow copy means slicing and deep copy means copy() 
2) b = a
   → Assignment (no copy) i.e., they are same array object point to same data
3) b = a[:] or b = a[1:] etc..
   → Shallow copy / View
   → two different array objects a and b point to same data
4) b = a.copy()
   → Deep copy
   → two different array objects a and b point to two diffrent array copies 
5) b+=10 does not create new array bt b = b+10 creates new array b
"""
a = np.arange(1,11)
print(a)
b = a
print(b)
b +=10
#b=b+10
#b[2]=22
print(b)
print(a)
print("================================")
a = np.arange(1,11)
c = a[1:8]
print(c)
print(a)
c[3] = 333
print(c)
print(a)
print("================================")
a = np.arange(1,11)
d = a.copy()
print(a)
print(d)
d[3]=123
print(d)
print(a)
print("================================")
```

```python
#Matrix operations
A = np.arange(1,7).reshape(3,2)
B = np.arange(6,12).reshape(2,3)
print(A)
print(B)
print(A@B)
print(np.dot(A,B))
print(A.T)
B.T
```

```python
# Stacking array(we use vectors or 1d arrays only)
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
print(np.vstack((a,b)))
print(np.hstack((a,b)))
print(np.column_stack((a,b)))
```

```python
# splitting array 
arr = np.arange(1,17).reshape(4,4)
print(arr)
print(np.vsplit(arr,2))
print(np.vsplit(arr,4))
print(np.hsplit(arr,2))
print(np.hsplit(arr,2))
```
